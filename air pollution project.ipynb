```python
df.drop(columns=["Date"],inplace=True)
```

```python
# Check the data types of each column
print(df.dtypes)

# Check for non-numeric columns
non_numeric_columns = df.select_dtypes(include=['object']).columns
print(f"Non-numeric columns: {non_numeric_columns}")
```

```python
# One-hot encode categorical columns (if any)
df = pd.get_dummies(df, columns=non_numeric_columns, drop_first=True)

# Verify the changes
print(df.head())
```

```python
# Assuming your target column is called 'TargetColumn', replace it with the actual name
X = df.drop('CO(GT)', axis=1)
y = df['CO(GT)']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the training data, and transform the test data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
df.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error  # Import mean_absolute_error

# Initialize the model
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
model.fit(X_train_scaled, y_train)

# Predict on the test data
y_pred = model.predict(X_test_scaled)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)  # Calculate mean absolute error

# Print the results
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
print(f'Mean Absolute Error: {mae}')  # Correct the print statement here
```
